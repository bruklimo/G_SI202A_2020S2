###Connect 4
Cada jogador coloca uma de suas peças em qualquer abertura no topo da grade. Jogam alternadamente até que um dos jogadores consigam alinhar 4 peças da mesma cor. A linha pode ser formada na vertical, na horizontal ou na diagonal.


#Tabuleiro 
O tabuleiro possui dimensões 6x7, 6 unidades de altura e 7 unidades de comprimento. A imagem mostra um tabuleiro vazio.
 tabuleiro vazio.png

##As dimensões to tabuleiro
O tabuleiro original possui dimensões de 6x7. Se o tabuleiro for aumentado ainda é possível jogar mas o tempo de jogo pode aumentar. O tabuleiro também pode ser diminuído mas isso afetaria mais o jogo, diminuir para 4x4 pode tornar a chance de empate maior, diminuir o tabuleiro para 3x4 ou 4x3 vai fazer com que apenas formações horizontais ou verticais sejam possíveis e a chance de empate muito maior e diminuir para 3x3 ou menos impossibilita o jogo como ele é, mas ainda pode ser possível se diminuir o critério da vitória de uma linha de 4 peças por uma linha de 3 peças, mas isso faria o jogo deixar de ser connect 4 para ser connect 3.


#Movimentos possíveis
Uma peça pode ser colocada no topo de qualquer pilha que ainda tenha espaço, ou seja, tenha menos de 6 peças. Uma vez colocada, a peça irá cair até o espaço mais baixo disponível.
